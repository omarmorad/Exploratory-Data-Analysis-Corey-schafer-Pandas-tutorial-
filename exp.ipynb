```python
Country_respondents = df['Country'].value_counts()
Country_respondents
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
""" we will do it using concatenation in the column axis"""
python_Usage = pd.concat([Country_respondents , country_uses_python] , axis = "columns" , sort = False )
python_Usage
```

```python

#Transform to a meaningful column names
python_Usage.rename(columns={'Country' : 'NumRespondents' , 'LanguageWorkedWith' : 'Numknowspython'}, inplace=True)
python_Usage
```

```python
#getting percentage
#(part/whole) * 100
python_Usage['PctKnowsPython'] = (python_Usage['Numknowspython']/python_Usage['NumRespondents']) * 100
python_Usage
```

```python
python_Usage.sort_values(by='PctKnowsPython', ascending=False , inplace=True)
python_Usage
```
